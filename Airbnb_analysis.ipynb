```python
Host_data= {'_id':[],'street':[],'suburb':[],
              'government_area':[],'market':[],'country':[],
              'country_code':[],'location_type':[], "longitude":[], "latitude":[],
                "is_location_exact":[]}

for i in host_pre["_id"]:
    Host_data["_id"].append(i)

for i in host_pre["address"]:
    Host_data['street'].append(i['street'])
    Host_data['suburb'].append(i["suburb"])
    Host_data['government_area'].append(i["government_area"])
    Host_data['market'].append(i["market"])
    Host_data['country'].append(i["country"])
    Host_data['country_code'].append(i["country_code"])
    
    Host_data['location_type'].append(i["location"]["type"])
    Host_data["longitude"].append(i["location"]["coordinates"][0])
    Host_data["latitude"].append(i["location"]["coordinates"][1])
    Host_data["is_location_exact"].append(i["location"]["is_location_exact"]) 
```

```python
def check_empty_values_improved(df):
    columns_to_check = [
        '_id', 'street', 'suburb', 'government_area', 'market',
        'country', 'country_code', 'location_type', 'longitude',
        'latitude', 'is_location_exact'
    ]
    
    empty_indices = {}
    
    for column in columns_to_check:
        empty_indices[column] = df[df[column] == ''].index.tolist()
    
    # New code to count empty values in each column
    print("Number of empty values in each column:")
    for column, indices in empty_indices.items():
        print(f"{column}: {len(indices)}")

    return empty_indices
```

AVAILABILITY

```python
#AVAILABILITY OF THE HOUSES IN THE NEIGHBOURHOODS

Availability = []

for i in collection.find({},{"_id":1, "availability":1}):
    Availability.append(i)

df_availability= pd.DataFrame(Availability)
df_availability
```

```python
df_availability_1 = {
    '_id': [],
    'availability_30': [],
    'availability_60': [],
    'availability_90': [],
    'availability_365': []
}

# Populate the '_id' field
df_availability_1['_id'] = df_availability['_id'].tolist()

# Populate the availability fields
# Note the correction from 'availabilty' to 'availability'
for availability in df_availability['availability']:
    df_availability_1['availability_30'].append(availability['availability_30'])
    df_availability_1['availability_60'].append(availability['availability_60'])
    df_availability_1['availability_90'].append(availability['availability_90'])
    df_availability_1['availability_365'].append(availability['availability_365'])

# Create the new DataFrame
Availability_df = pd.DataFrame(df_availability_1)
```

```python
Availability_df #(FINAL)
```

```python
Amenities = []
for i in collection.find({},{"_id":1, "amenities":1}):
    Amenities.append(i)

df_Amenities= pd.DataFrame(Amenities)
df_Amenities
```

```python
def sort_amenities(x):
    a= x
    a.sort()
    return a


df_Amenities["amenities"]= df_Amenities["amenities"].apply(lambda x: sort_amenities(x))
df_Amenities # DF with sorted amenities(FINAL)
```

JOIN ALL DATA FRAME 

```python
df_merged = df_viz.copy()

df_merged = pd.merge(df_merged, df_viz_h, on='_id')
df_merged = pd.merge(df_merged, Host_data_neighbourhood, on='_id')
df_merged = pd.merge(df_merged, Availability_df, on='_id')
df_merged = pd.merge(df_merged, df_Amenities, on='_id')

df_merged
```

```python
# covert to csv
df_merged.to_csv("Airbnb.csv", index=False)

# save csv file
```
