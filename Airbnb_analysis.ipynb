```python
def check_empty_values_improved(df):
    columns_to_check = [
        '_id', 'street', 'suburb', 'government_area', 'market',
        'country', 'country_code', 'location_type', 'longitude',
        'latitude', 'is_location_exact'
    ]
    
    empty_indices = {}
    
    for column in columns_to_check:
        empty_indices[column] = df[df[column] == ''].index.tolist()
    
    # New code to count empty values in each column
    print("Number of empty values in each column:")
    for column, indices in empty_indices.items():
        print(f"{column}: {len(indices)}")

    return empty_indices
```

```python
#AVAILABILITY OF THE HOUSES IN THE NEIGHBOURHOODS

Availability = []

for i in collection.find({},{"_id":1, "availability":1}):
    Availability.append(i)

df_availability= pd.DataFrame(Availability)
df_availability
```

```python
df_availability_1 = {
    '_id': [],
    'availability_30': [],
    'availability_60': [],
    'availability_90': [],
    'availability_365': []
}

# Populate the '_id' field
df_availability_1['_id'] = df_availability['_id'].tolist()

# Populate the availability fields
# Note the correction from 'availabilty' to 'availability'
for availability in df_availability['availability']:
    df_availability_1['availability_30'].append(availability['availability_30'])
    df_availability_1['availability_60'].append(availability['availability_60'])
    df_availability_1['availability_90'].append(availability['availability_90'])
    df_availability_1['availability_365'].append(availability['availability_365'])

# Create the new DataFrame
Availability_df = pd.DataFrame(df_availability_1)
```

```python
import plotly.express as px
import streamlit as st
import pandas as pd

# Assuming df is your DataFrame
# First, we need to preprocess the amenities data
def process_amenities(df):
    # Split the amenities string into a list
    df['amenities_list'] = df['amenities'].str.strip('{}').str.split(',')
    
    # Explode the DataFrame so each amenity gets its own row
    exploded_df = df.explode('amenities_list')
    
    # Clean up the amenity names
    exploded_df['amenities_list'] = exploded_df['amenities_list'].str.strip().str.strip('"')
    
    # Group by amenity and calculate average price
    amenity_prices = exploded_df.groupby('amenities_list')['price'].mean().sort_values(ascending=False).reset_index()
    
    return amenity_prices.head(20)  # Return top 20 amenities by average price

# Process the data
amenity_prices = process_amenities(df)

# Create the bar chart
fig5 = px.bar(amenity_prices,
              x='amenities_list',
              y='price',
              title='Average Price for Top 20 Amenities',
              labels={'amenities_list': 'Amenity', 'price': 'Average Price'},
              height=600)

# Customize the layout
fig5.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='',
    yaxis_title='Average Price ($)',
    margin=dict(b=100)  # Increase bottom margin to prevent label cutoff
)

# Display the chart in Streamlit
fig5.show()
```
